```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pd.pandas.set_option('display.max_columns',None)

df = pd.read_csv("train.csv")
```

```python
df.head()
```

```python
df.drop(columns=["Id"],inplace=True)
df.shape
```

```python
df.describe()
```

```python
# Missing values
missing_values_features = [feature for feature in df.columns if df[feature].isnull().sum() > 1]
missing_values_features

for feature in missing_values_features:
    print(feature ,df[feature].isnull().sum(),",", np.round(df[feature].isnull().mean(),4), "%", "missing values" )
```

```python
# Since missing values are more finding relation between missing value features and sales price.
for missing_val_features in missing_values_features:
    
    # Make a copy of data
```

```python
data = df.copy()

# replace nan vales with 1 and othes with 0



for missing_val_features in missing_values_features:
    
    data[missing_val_features] = np.where(data[missing_val_features].isnull(),1,0)
    grp = data.groupby(missing_val_features).mean()
    display(grp)
    grp['SalePrice'].plot.bar()
    plt.show()
    
    
    
```

From the above graphs we can conclude that we have to replace Nan values with some value.

```python
# Numerical Variabels and relation with salesprice

numerical_features = [feature for feature in df.columns if df[feature].dtypes != 'O']
print("Number of numerical variables :", len(numerical_features))

df[numerical_features].head()

data = df.copy()

for num_feature in numerical_features:
    display(data.groupby(num_feature).mean())
    data.groupby(num_feature)['SalePrice'].mean().plot()
    plt.xlabel(num_feature)
    plt.ylabel('Mean House Price')
    plt.show()
    
```

```python
# Let's find the relation between the age of house and the salesprice

year_features=[feature for feature in df.columns if 'Yr' in feature or 'Year' in feature]
print(year_features)
```

```python
# age of house, remodel house age, garage of house age = 'YearBuilt' or 'GarageYrBlt' or 'YearRemodAdd'  - 'YrSold'

for feature in year_features:
    
    if feature != 'YrSold':
        data  = df.copy()
        
        data[feature] = data['YrSold'] - data[feature]
        
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature+" age")
        plt.ylabel("SalePrice")
        plt.show()
    
```
